# III. Methods & Results

3 kinds of deep neural networks were implemented using Tensorflow. Each had 6 hidden layers, and each hidden layer up through the 5th used a sigmoid activation function. The 6th hidden layer uses a Rectified Linear Unit (ReLU) activation function as it's outputted matrix is multiplied by the output layer in which the class with the highest output value is the predicted class. This combination was useful because using only sigmoid activation might have resulted in a vanishing gradient while the ReLU activation tends to blow it up.

The graphs have been included under each deep neural network as they are produced immediately after the training is finished.

### First Deep Neural Network

There are 30 inputted features (all of which are standardized on a scale of 0-1 per respective feature). The first deep neural network is a feed forward network that alternates in the number of nodes per hidden layer between double the number of inputs and that number. So, the number of nodes per hidden layer went as follows: 60 -> 30 -> 60 -> 30 -> 60 -> 30. This structure was chosen because there were many features in the data set, the alternating structure would likely yield better results (or would be more effective than a deep neural network with 30 nodes per layer), and because it didn't put to much emphasis on any one part of the network.

The code below takes the data, shuffles it to a consistent random state, divides it into a training and test set (20%), and converts each y output into a list of size 2 in which a 1 is placed in the first index if it is benign or in the second index if it is malignant. The learning rate is set to 0.05 for consistency through each of the 3 networks, and the number of epochs is limited to 200 as this resulted in the highest accuracy in both the training and test sets by avoiding overfitting (using the early stop technique). For each epoch, the mean squared error and accuracy are both calculated for the training set, and the mean squared error and accuracy are finally calculated for the test set. 

_The code was implemented after I watched this tutorial from edureka https://www.youtube.com/watch?v=yX8KuPZCAMo_

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split


def one_hot_encode(labels): #Matrix with 1 in the index corresponding to the proper class
    n_labels = len(labels)
    n_unique_labels = len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels, n_unique_labels))
    one_hot_encode[np.arange(n_labels), labels] = 1
    return one_hot_encode

def deep_feed_forward(x, weights, biases):
    #Hidden layer 1 with sigmoid activation
    layer_1 = tf.add(tf.matmul(x, weights['h1']),  biases['b1']) #matrix multiplication, then adding of biases
    layer_1 = tf.nn.sigmoid(layer_1)
    
    #Hidden layer 2 with sigmoid activation
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']),  biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    
    #Hidden layer 3 with sigmoid activation
    layer_3 = tf.add(tf.matmul(layer_2, weights['h3']),  biases['b3'])
    layer_3 = tf.nn.sigmoid(layer_3)
    
    #Hidden layer 4 with sigmoid activation
    layer_4 = tf.add(tf.matmul(layer_3, weights['h4']),  biases['b4'])
    layer_4 = tf.nn.sigmoid(layer_4)
    
    #Hidden layer 5 with sigmoid activation
    layer_5 = tf.add(tf.matmul(layer_4, weights['h5']),  biases['b5'])
    layer_5 = tf.nn.sigmoid(layer_5)
    
    #Hidden layer 6 with relu activation
    layer_6 = tf.add(tf.matmul(layer_5, weights['h6']),  biases['b6'])
    layer_6 = tf.nn.relu(layer_6)
    
    #Output layer with linear activation
    out_layer = tf.matmul(layer_6, weights['out']) + biases['out']
    return out_layer
```

```python
data = pd.read_csv('BreastCancerWisconsin.csv')
X = data[data.columns[1:32]].values
y = data[data.columns[0]]
X = preprocessing.MinMaxScaler().fit_transform(X) #Normalizes X features (scales to range of 0-1)
original_X = X
original_y = y

encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
Y = one_hot_encode(y) #[0, 1] is Malignant; [1, 0] is Benign

X, Y = shuffle(X, Y, random_state = 1)
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.20, random_state = 415)

learning_rate = 0.05
training_epochs = 200
cost_history = np.empty(shape = [1], dtype = float)
n_dim = X.shape[1] #Number of columns
n_class = 2 #number of classes

#30 features to be used as inputs #DEEP FEED FORWARD NETWORK BELOW
n_hidden_1 = 60 #60 nodes in hidden layer 1
n_hidden_2 = 30
n_hidden_3 = 60
n_hidden_4 = 30
n_hidden_5 = 60
n_hidden_6 = 30

x = tf.placeholder(tf.float32, [None, n_dim]) #Latter input is shape of input; None indicates that as many rows (samples) as you like can be used
W = tf.Variable(tf.zeros([n_dim, n_class])) #Weight matrix
b = tf.Variable(tf.zeros([n_class]))
y_ = tf.placeholder(tf.float32, [None, n_class]) #outputs; None indicates as many rows (samples) as you like. n_class indicates number of columns (classes)

weights = {
    'h1': tf.Variable(tf.truncated_normal([n_dim, n_hidden_1])), #Matrix is number of features by number of nodes in hidden layer 1
    'h2': tf.Variable(tf.truncated_normal([n_hidden_1, n_hidden_2])),
    'h3': tf.Variable(tf.truncated_normal([n_hidden_2, n_hidden_3])),
    'h4': tf.Variable(tf.truncated_normal([n_hidden_3, n_hidden_4])),
    'h5': tf.Variable(tf.truncated_normal([n_hidden_4, n_hidden_5])),
    'h6': tf.Variable(tf.truncated_normal([n_hidden_5, n_hidden_6])),
    'out': tf.Variable(tf.truncated_normal([n_hidden_6, n_class]))
}

biases = {
    'b1': tf.Variable(tf.truncated_normal([n_hidden_1])),
    'b2': tf.Variable(tf.truncated_normal([n_hidden_2])),
    'b3': tf.Variable(tf.truncated_normal([n_hidden_3])),
    'b4': tf.Variable(tf.truncated_normal([n_hidden_4])),
    'b5': tf.Variable(tf.truncated_normal([n_hidden_5])),
    'b6': tf.Variable(tf.truncated_normal([n_hidden_6])),
    'out': tf.Variable(tf.truncated_normal([n_class])) #2 classes
}
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver() #saves model

#Call model
y = deep_feed_forward(x, weights, biases)

#Define cost function and optimizer
cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = y, labels = y_))  #Loss function; logits is output, y_ is target values (already known)
training_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost_function) #Gradient descent with learning rate; minimizing loss (cost_function)

sess = tf.Session()
sess.run(init)

#Calculate cost and accuracy for each epoch
mse_history = []
accuracy_history = []

for epoch in range(training_epochs):
    sess.run(training_step, feed_dict={x: train_x, y_: train_y})
    cost = sess.run(cost_function, feed_dict={x: train_x, y_: train_y})
    cost_history = np.append(cost_history, cost) #Appends cost
    correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    pred_y = sess.run(y, feed_dict={x: test_x})
    mse = tf.reduce_mean(tf.square(pred_y - test_y))
    mse_ = sess.run(mse)
    mse_history.append(mse_)
    accuracy = (sess.run(accuracy, feed_dict={x: train_x, y_: train_y}))
    accuracy_history.append(accuracy)
```

```python
#Plot mse and accuracy graph
print('MSE history')
plt.plot(mse_history, 'r')
plt.show()
print('Accuracy over epochs')
plt.plot(accuracy_history)
plt.show()

#Print the final accuracy
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print("Test accuracy: ", (sess.run(accuracy, feed_dict={x: test_x, y_: test_y})))

#Print the final mean square error
pred_y = sess.run(y, feed_dict={x: test_x})
mse = tf.reduce_mean(tf.square(pred_y - test_y))
print("MSE: %.4f" % sess.run(mse))

sess.close()
```

The results above show that this structure was very effective in developing a model that could predict the malignancy of a tumor based on the nuclei of the cells in it. Over the epochs, the MSE decreased significantly up until ~65th-70th epoch, where it rose again, slightly at ~130. The accuracy on the training set oscillated a lot up until around that epoch, after which it approached the 90% mark at ~75th-80th epoch and leveled off at around 95% accuracy.

The test accuracy of 95.614% and MSE of 130.8046 show that overfitting was avoided due to the early stop technique and that the model performed reasonably well this time on new data.

### Second Deep Neural Network

The second deep neural network is a feed forward network that begins with a hidden layer with twice as many nodes as inputted features that dropped down by 10 each time until finally being passed on to the output layer. So, the number of nodes per hidden layer went as follows: 60 -> 50 -> 40 -> 30 -> 20 -> 10. This structure was chosen to examine the effects of a contracting neural network. It places less emphasis at the beginning of the network due to the far greater number of nodes in the hidden layers there and more emphasis on the end due to the fewer nodes in the latter hidden layers.

The code below performs the same actions as in the first deep neural network. The only difference is in the number of nodes per hidden layer.

```python
data = pd.read_csv('BreastCancerWisconsin.csv')
X = data[data.columns[1:32]].values
y = data[data.columns[0]]
X = preprocessing.MinMaxScaler().fit_transform(X) #Normalizes X features (scales to range of 0-1)
original_X = X
original_y = y

encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
Y = one_hot_encode(y) #[0, 1] is Malignant; [1, 0] is Benign

X, Y = shuffle(X, Y, random_state = 1)
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size = 0.20, random_state = 415)

learning_rate = 0.05
training_epochs = 200
cost_history = np.empty(shape = [1], dtype = float)
n_dim = X.shape[1] #Number of columns
n_class = 2 #number of classes

#30 features to be used as inputs #DEEP FEED FORWARD CONTRACTING (Getting narrower) NETWORK BELOW
n_hidden_1 = 60 #60 nodes in hidden layer 1
n_hidden_2 = 50
n_hidden_3 = 40
n_hidden_4 = 30
n_hidden_5 = 20
n_hidden_6 = 10

x = tf.placeholder(tf.float32, [None, n_dim]) #Latter input is shape of input; None indicates that as many rows (samples) as you like can be used
W = tf.Variable(tf.zeros([n_dim, n_class])) #Weight matrix
b = tf.Variable(tf.zeros([n_class]))
y_ = tf.placeholder(tf.float32, [None, n_class]) #outputs; None indicates as many rows (samples) as you like. n_class indicates number of columns (classes)

weights = {
    'h1': tf.Variable(tf.truncated_normal([n_dim, n_hidden_1])), #Matrix is number of features by number of nodes in hidden layer 1
    'h2': tf.Variable(tf.truncated_normal([n_hidden_1, n_hidden_2])),
    'h3': tf.Variable(tf.truncated_normal([n_hidden_2, n_hidden_3])),
    'h4': tf.Variable(tf.truncated_normal([n_hidden_3, n_hidden_4])),
    'h5': tf.Variable(tf.truncated_normal([n_hidden_4, n_hidden_5])),
    'h6': tf.Variable(tf.truncated_normal([n_hidden_5, n_hidden_6])),
    'out': tf.Variable(tf.truncated_normal([n_hidden_6, n_class]))
}

biases = {
    'b1': tf.Variable(tf.truncated_normal([n_hidden_1])),
    'b2': tf.Variable(tf.truncated_normal([n_hidden_2])),
    'b3': tf.Variable(tf.truncated_normal([n_hidden_3])),
    'b4': tf.Variable(tf.truncated_normal([n_hidden_4])),
    'b5': tf.Variable(tf.truncated_normal([n_hidden_5])),
    'b6': tf.Variable(tf.truncated_normal([n_hidden_6])),
    'out': tf.Variable(tf.truncated_normal([n_class])) #2 classes
}
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver() #saves model

#Call model
y = deep_feed_forward(x, weights, biases)

#Define cost function and optimizer
cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = y, labels = y_))  #Loss function; logits is output, y_ is target values (already known)
training_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost_function) #Gradient descent with learning rate; minimizing loss (cost_function)

sess = tf.Session()
sess.run(init)

#Calculate cost and accuracy for each epoch
mse_history = []
accuracy_history = []

for epoch in range(training_epochs):
    sess.run(training_step, feed_dict={x: train_x, y_: train_y})
    cost = sess.run(cost_function, feed_dict={x: train_x, y_: train_y})
    cost_history = np.append(cost_history, cost) #Appends cost
    correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    pred_y = sess.run(y, feed_dict={x: test_x})
    mse = tf.reduce_mean(tf.square(pred_y - test_y))
    mse_ = sess.run(mse)
    mse_history.append(mse_)
    accuracy = (sess.run(accuracy, feed_dict={x: train_x, y_: train_y}))
    accuracy_history.append(accuracy)
```

The test accuracy of 96.0526% and MSE of 4.3721 reveal that this model also doesn't suffer from overfitting and is also very effective in classifying unseen samples.
